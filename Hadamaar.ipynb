```python
def walsh(x):
    
    #x = np.array(x)
    if(len(x.shape)<2): # make sure n is 1D
        if(len(x)>3): # make sure M = 2
            n = len(x)
            m = trunc(log(n,2))
            x = x[0:2**M]
            H2 = np.array([[1,1],[1,-1]])
            for i in range(M-1):
                if i==0:
                    H = np.kron(H2,H2)
                else:
                    H = np.kron(H,H2)
                    
            return (np.dot(H,x)/2.**M,x, M)
        else:
            print("array too short")
    else:
        print("1d array expected")
    
```

```python
def hada(N):
    H = np.zeros((1,1), dtype=int)
    H[0] = 1
    log_N = int(log(N,2))
    shape = H.shape[0]
    while log_N>0:
        A = np.zeros((H.shape[0]*2,H.shape[0]*2),dtype=int)
        r1,c1 = 0,0
        r2,c2 = 0, H.shape[0]
        r3,c3 = H.shape[0],0
        r4,c4 = H.shape[0], H.shape[0]
        A[r1:r1+H.shape[0], c1:c1+H.shape[1]] = H
        A[r2:r2+H.shape[0], c2:c2+H.shape[1]] = H   
        A[r3:r3+H.shape[0], c3:c3+H.shape[1]] = H
        A[r4:r4+H.shape[0], c4:c4+H.shape[1]] = -H
        H = A
        log_N -= 1
    return H
    
    
```
